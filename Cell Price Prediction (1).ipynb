```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
import pickle

# 1. Read data
dataset = pd.read_csv("datasets_11167_15520_train.csv")
print(dataset.head())

# 2. Features (all except price_range)
X = np.array(dataset.drop(['price_range'], axis=1), dtype=np.int64)
# Target = price_range
y = np.array(dataset['price_range'], dtype=np.int64)

# 3. Split data
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(
    X, y, test_size=0.1
)

# 4. Train model (Linear Regression, like your old code)
best = 0
for _ in range(1000):
    x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(
        X, y, test_size=0.1
    )

    linear = linear_model.LinearRegression()
    linear.fit(x_train, y_train)
    acc = linear.score(x_test, y_test)

    if acc > best:
        best = acc
        with open("studentmodel_v2.pickle", "wb") as f:  # save as new version
            pickle.dump(linear, f)

print("✅ Best Accuracy:", best)

# 5. Load saved model
with open("studentmodel_v2.pickle", "rb") as f:
    newlinear = pickle.load(f)

# 6. Example input (must match dataset features count = 20)
# [battery_power, blue, clock_speed, dual_sim, fc, four_g, int_memory,
#  m_dep, mobile_wt, n_cores, pc, px_height, px_width, ram,
#  sc_h, sc_w, talk_time, three_g, touch_screen, wifi]
mine = [[842,1,2,0,1,0,7,0,188,2,1,20,756,2549,9,7,19,0,1,0]]
mine = np.array(mine)

print("🔮 Prediction for mine:", newlinear.predict(mine))

# 7. Predictions on test data
predictions = newlinear.predict(x_test)

for i in range(10):  # show only first 10 for readability
    print(f"Predicted: {predictions[i]:.2f}, Features: {x_test[i]}, Actual: {y_test[i]}")
```
